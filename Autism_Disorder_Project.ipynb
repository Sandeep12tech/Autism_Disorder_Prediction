# Feature Engineering

```python
df.columns
```

```python
df["age"].unique()
```

```python
def agegroup(age):
    if age<4:
        return "Toddler"
    elif age<12:
        return "Kid"
    elif age<18:
        return "Teenager"
    elif age<40:
        return "Young"
    else:
        return "Senior"
        
```

```python
df["ageGroup"] = df["age"].map(agegroup)
```

```python
df.head(3)
```

```python
plt.figure(figsize=(4,2))
sns.countplot(x=df["ageGroup"],hue = df["Class/ASD"])
plt.show()
```

```python
def add_feature(data):
    data["sum_score"] = 0
    for col in data.loc[:,"A1_Score":"A10_Score"].columns:
        data["sum_score"]=data["sum_score"] + data[col]
    return data
```

```python
df.info()
```

```python
df = add_feature(df)
```

```python
plt.figure(figsize=(4,2))
sns.countplot(x=df["sum_score"],hue = df["Class/ASD"])
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.distplot(df["age"])
plt.show()
```

```python
def encode_labels(data):
    for col in data.columns:
        if(data[col].dtype == "object"):
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
    return data
```

```python
df = encode_labels(df)
```

# Model Training

```python
df.info()
```

```python
X = df.drop(columns=["ID","austim","used_app_before","age_desc","Class/ASD"])
```

```python
y = df["Class/ASD"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 10,test_size=0.2)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
models = [SVC() , LogisticRegression()]
```

```python
for model in models:
    model.fit(X_train,y_train)
    print(model,": Training Accuracy : ",metrics.roc_auc_score(y_train,model.predict(X_train)))
    print(model,": Validation Accuracy : ",metrics.roc_auc_score(y_test,model.predict(X_test)))
```
